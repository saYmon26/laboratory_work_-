```python
with open('lab2.txt', 'r') as file:
    lines = file.readlines()
    non_empty_lines = [line for line in lines if line.strip()]
    for line in non_empty_lines:
        print(line, end='') 
```

```python
lines_list = []  
with open('lab2.txt', 'r') as file:
    for line in file:
        lines_list.append(line.strip()) 

    print(lines_list)
```
