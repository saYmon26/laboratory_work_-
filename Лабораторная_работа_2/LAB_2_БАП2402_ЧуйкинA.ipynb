```python
list_full = []
with open('log_git.txt', 'r') as file:
    for line in file:
        list_full.append(line)

target_author = "Ivan Ivanov <ivan@example.com>"
ivan_commits = []

list_n0 = list_full.copy()
content = 0

while content < 26 and len(list_n0) >= 5:
    list_n = list_n0[:5] 
    
    if len(list_n) >= 5:
        author_line = list_n[1].strip()
        message_line = list_n[4].strip()
        
        if author_line == f"Author: {target_author}":
            ivan_commits.append(message_line)
            print(f"Найден коммит: {message_line}")
    
    if len(list_n0) >= 6:
        del list_n0[:6]
    else:
        break
    
    content += 1

print(f"\nВсего коммитов у {target_author}: {len(ivan_commits)}")
```
