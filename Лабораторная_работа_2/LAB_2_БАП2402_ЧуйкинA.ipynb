```python
import json

with open('devices.json', 'r') as file:
    data = json.load(file)

for device in data['devices']:
    name = device['name']
    brand = device['specifications']['brand']
    dimensions = device['specifications']['dimensions']
    
    size_parts = []
    for key, value in dimensions.items():
        size_parts.append(f"{key}: {value}")
    size = ", ".join(size_parts)
    
    print(name)
    print(brand)
    print(size)
```
