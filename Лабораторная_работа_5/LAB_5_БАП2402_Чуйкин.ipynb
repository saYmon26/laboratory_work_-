```python
import requests

def get_posts(user_id):
    url = f"https://jsonplaceholder.typicode.com/posts?userId={user_id}"
    response = requests.get(url)
    posts = response.json()
    
    print(f"Первые 5 постов пользователя {user_id}:")
    for i in range(5):
        print(f"{i+1}. {posts[i]['title']}")
    return posts


get_posts(1)
```
